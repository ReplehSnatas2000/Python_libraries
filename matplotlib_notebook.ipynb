```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
%matplotlib inline
```

```python
l1 = np.linspace(0, 5, 11)
y = l1**2
y
```

```python
plt.plot(l1, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.subplot(1, 2, 1)
plt.plot(l1, y, "r")

plt.subplot(1, 2, 2)
plt.plot(y, l1, "b")
```

```python
# Object oriented method
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(l1, y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(l1, y)
axes2.plot(l1, y)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2)
# fig, axes = plt.subplots(nrows=3, ncols=2)
# for x in axes:
#     for x1 in x:
#         x1.plot(l1,y)
axes[1][0].plot(l1, y)
plt.tight_layout()
axes
```

```python
fig, axes3 = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))
axes3[0].plot(l1, y)
axes3[1].plot(y, l1)
plt.tight_layout()
```

```python
fig.savefig("my_plot.jpg")
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(l1, l1**2, label="Square")
ax.plot(l1, l1**3, label="Cube")
ax.legend(loc=(0.5, 0.5))
```

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0, 0, 1, 1])
ax1.plot(l1, y, color="red", linewidth=5,
         alpha=0.5, linestyle="-.", marker="o", markersize=10, markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="black")
```

```python
l2 = np.arange(-5, 5)
fig2 = plt.figure()
ax2 = fig2.add_axes([0, 0, 1, 1])
y1 = np.array(list(map(lambda x: math.exp(x), l2)))
ax2.plot(l2, y1)
ax2.set_xlim([1, 4])
```
