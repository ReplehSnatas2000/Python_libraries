```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
print(tips["total_bill"].max())
tips
```

```python
sns.distplot(tips["total_bill"], kde=False, bins=14)
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
```

```python
sns.displot(tips["size"])
```

```python
sns.pairplot(tips, hue="sex")
```

```python
sns.pairplot(tips, hue="smoker")
```

```python
sns.rugplot(tips["total_bill"])
```

```python
sns.barplot(x="sex", y="total_bill", data=tips, estimator=np.std)
```

```python
sns.countplot(x="sex", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(ata=tips,x="day", y="total_bill", jitter=False)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, color="black")
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
tc = tips.corr()
# print(tc)
sns.heatmap(tc, annot=True)
```

```python
flights = sns.load_dataset("flights")
mat = flights.pivot_table(index="month", columns="year", values="passengers")
mat
```

```python
sns.heatmap(mat)
```

```python
sns.clustermap(mat)
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips, row="smoker", col="time")
g.map(sns.histplot, "total_bill", "tip")
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, markers="+", hue="sex")
sns.lmplot(x="tip", y="total_bill", data=tips, markers="+")
```

```python
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=3)
sns.countplot(x="smoker", data=tips)
sns.despine()
tips[tips["smoker"] == "Yes"].count()
tips[tips["smoker"] == "No"].count()
```
